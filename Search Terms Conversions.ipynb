# Search Terms Conversions

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(filepath_or_buffer="D:/ALEXANDER/Downloads/search_terms_july_2024.csv", encoding='latin1')
```

```python
df.head(2)
```

```python
x = df['Termino de busqueda'].str.replace('Ã³','ó').str.replace('Ã¡','á').str.replace('Ã','í').str.replace('í©','í')

df = df[['Impr.', 'Interacciones', 'Conversiones']]
df.head()
```

```python
df2 = pd.concat([x,df], axis = 1)

df2 = df2.rename(columns={"Interacciones":"Clicks",
                        "Impr.":"Impresiones"})
df2.head()
```

```python
conv = df2['Conversiones'].str.replace(',','.')

df2 = df2.drop(columns=['Conversiones'])
df2.head()
```

```python
df2 = pd.concat([df2, conv], axis = 1)
df2.head()
```

```python
def convert_to_float(value):
    if isinstance(value, str):
        return float(value.replace(',', '.'))
    else:
        return float(value)
```

```python
df_num = df2[['Impresiones','Clicks','Conversiones']].applymap(convert_to_float)
df_num.head()
```

```python
search_terms = df2['Termino de busqueda']
df_grouped = pd.concat([search_terms, df_num], axis = 1)
df_grouped.head()
```

```python
df_grouped2 = df_grouped.groupby('Termino de busqueda')[['Impresiones', 'Clicks', 'Conversiones']].sum().reset_index()
df_grouped2.head()
```

```python
df_grouped2 = df_grouped2.sort_values(by='Conversiones', ascending = False)
df_grouped2.head(2)
```

```python
df_final = df_grouped2[df_grouped2['Conversiones']>=0.1]
df_final.info
```

```python
df_final.describe()
```

```python
df_final.to_csv('Portfolio_Search_Terms3.csv', index=False,encoding='latin1')
```
